# Basic visualization of stock price data
In this Notebook some basic visualization for stock prices are displayed.

#### Importing needed libraries.

```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
```

#### Getting stock data from Yahoo.

```python
start = pd.to_datetime('2015-01-01')
neste = web.DataReader('NESTE.HE', 'yahoo', start=start)
sampo = web.DataReader('SAMPO.HE', 'yahoo', start=start)
wartsila = web.DataReader('WRT1V.HE', 'yahoo', start=start)
```

#### Plotting the closing prices.

```python
plt.style.use('seaborn')
neste['Close'].plot.line(label = 'Neste', title = "Closing Prices", figsize=(14,8) )
sampo['Close'].plot.line(label = 'Sampo')
wartsila['Close'].plot.line(label = 'Wrtsil')
plt.legend();
```

This picture seems to show that Sampo and Wrtsil has not been so good investments. Lets plot still the adjusted closing prices because they take into account any corporate actions like dividens.

```python
neste['Adj Close'].plot.line(label = 'Neste', title = "Adjustet Closing Prices", figsize=(14,8) )
sampo['Adj Close'].plot.line(label = 'Sampo')
wartsila['Adj Close'].plot.line(label = 'Wrtsil')
plt.legend();
```

The picture didn't change a much. Well, it looks like during that time period even when taking into account the dividends, Neste was a really good investment, with Sampo gains depend on when did you buy it, and Wrtsil was ok until 2018.
